```python
cust_item = all_df2.pivot_table(index='CustomerID', columns='StockCode', values ='Quantity',aggfunc='sum')
```

```python
cust_item1 = cust_item.applymap(lambda x : 1 if x>0 else 0)
cust_item1.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cust_sim = pd.DataFrame(cosine_similarity(cust_item1))
```

```python
def get_items_to_recommend_cust(cust_a):
  '''returns the items to recommend to a customer using customer similarity'''
  most_similar_user = cust_sim1.loc[cust_a].sort_values(ascending=False).reset_index().iloc[1, 0]
  items_bought_by_cust_a = set(cust_item1.loc[cust_a].iloc[cust_item1.loc[cust_a].to_numpy().nonzero()].index)
  items_bought_by_cust_b = set(cust_item1.loc[most_similar_user].iloc[cust_item1.loc[most_similar_user].to_numpy().nonzero()].index)
  items_to_recommend_to_a = items_bought_by_cust_b - items_bought_by_cust_a
  items_description = all_df1.loc[all_df1['StockCode'].isin(items_to_recommend_to_a), ['StockCode', 'Description']].drop_duplicates().set_index('StockCode')
  return items_description
```
